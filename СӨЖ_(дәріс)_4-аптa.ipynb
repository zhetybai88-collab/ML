```python
print("=== 1Ô∏è‚É£ –û–±—ä–µ–∫—Ç—ñ–≥–µ –±–∞“ì—ã—Ç—Ç–∞–ª“ì–∞–Ω –±–∞“ì–¥–∞—Ä–ª–∞–º–∞–ª–∞—É (–û–û–ë) ===")

# üîπ –ö–ª–∞—Å—Å –∂”ô–Ω–µ –æ–±—ä–µ–∫—Ç –º—ã—Å–∞–ª—ã
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        """–¢—ñ–∫—Ç”©—Ä—Ç–±“±—Ä—ã—à—Ç—ã“£ –∞—É–¥–∞–Ω—ã"""
        return self.width * self.height

    def perimeter(self):
        """–¢—ñ–∫—Ç”©—Ä—Ç–±“±—Ä—ã—à—Ç—ã“£ –ø–µ—Ä–∏–º–µ—Ç—Ä—ñ"""
        return 2 * (self.width + self.height)

# üîπ –û–±—ä–µ–∫—Ç—ñ–ª–µ—Ä–¥—ñ “õ“±—Ä—É
r = Rectangle(5, 3)
print("–ê—É–¥–∞–Ω—ã:", r.area())
print("–ü–µ—Ä–∏–º–µ—Ç—Ä—ñ:", r.perimeter())

print("\n=== 2Ô∏è‚É£ –ï—Ä–µ–∫—à–µ –∂–∞“ì–¥–∞–π–ª–∞—Ä–¥—ã ”©“£–¥–µ—É (try-except) ===")

# üîπ “ö–∞—Ç–µ ”©“£–¥–µ—É –º—ã—Å–∞–ª—ã
try:
    number = int(input("–°–∞–Ω –µ–Ω–≥—ñ–∑—ñ“£—ñ–∑: "))
    result = 10 / number
except ZeroDivisionError:
    print("“ö–∞—Ç–µ: 0-–≥–µ –±”©–ª—É–≥–µ –±–æ–ª–º–∞–π–¥—ã!")
except ValueError:
    print("“ö–∞—Ç–µ: –¢–µ–∫ —Å–∞–Ω –µ–Ω–≥—ñ–∑—ñ“£—ñ–∑!")
else:
    print("–ù”ô—Ç–∏–∂–µ:", result)
finally:
    print("–ë–∞“ì–¥–∞—Ä–ª–∞–º–∞ –∞—è“õ—Ç–∞–ª–¥—ã.\n")


print("=== 3Ô∏è‚É£ –ú–æ–¥—É–ª—å–¥–µ—Ä –º–µ–Ω –ø–∞–∫–µ—Ç—Ç–µ—Ä ===")

# Colab-—Ç–∞ –º–æ–¥—É–ª—å–¥—ñ —Ñ–∞–π–ª —Ç“Ø—Ä—ñ–Ω–¥–µ –∂–∞—Å–∞–π–º—ã–∑
with open("mymath.py", "w") as f:
    f.write("""
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "0-–≥–µ –±”©–ª—É–≥–µ –±–æ–ª–º–∞–π–¥—ã"
""")

# üîπ –ï–Ω–¥—ñ –º–æ–¥—É–ª—å–¥—ñ –∏–º–ø–æ—Ä—Ç—Ç–∞–π–º—ã–∑
import mymath

print("“ö–æ—Å—É:", mymath.add(5, 3))
print("–ê–∑–∞–π—Ç—É:", mymath.sub(10, 7))
print("–ö”©–±–µ–π—Ç—É:", mymath.mul(4, 6))
print("–ë”©–ª—É:", mymath.div(8, 2))
print("–ë”©–ª—É (“õ–∞—Ç–µ):", mymath.div(8, 0))

print("\n=== 4Ô∏è‚É£ NumPy –∫—ñ—Ç–∞–ø—Ö–∞–Ω–∞—Å—ã–Ω–∞ –∫—ñ—Ä—ñ—Å–ø–µ ===")

import numpy as np

# üîπ –ë—ñ—Ä ”©–ª—à–µ–º–¥—ñ –º–∞—Å—Å–∏–≤
a = np.array([1, 2, 3, 4])
print("–ú–∞—Å—Å–∏–≤ a:", a)
print("–¢“Ø—Ä—ñ:", type(a))

# üîπ –ï–∫—ñ ”©–ª—à–µ–º–¥—ñ –º–∞—Å—Å–∏–≤
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print("\n–ú–∞—Å—Å–∏–≤ b:\n", b)
print("”®–ª—à–µ–º—ñ:", b.shape)
print("–≠–ª–µ–º–µ–Ω—Ç —Å–∞–Ω—ã:", b.size)
print("–ú–∞—Å—Å–∏–≤ —Ç“Ø—Ä—ñ:", b.dtype)

# üîπ –ù–µ–≥—ñ–∑–≥—ñ –æ–ø–µ—Ä–∞—Ü–∏—è–ª–∞—Ä
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print("\n“ö–æ—Å—É:", x + y)
print("–≠–ª–µ–º–µ–Ω—Ç—Ç—ñ–∫ –∫”©–±–µ–π—Ç—É:", x * y)
print("–°–∫–∞–ª—è—Ä–ª—ã“õ –∫”©–±–µ–π—Ç—ñ–Ω–¥—ñ:", x @ y)

# üîπ –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è –º—ã—Å–∞–ª—ã
numbers = np.arange(1, 6)
squared = numbers ** 2
print("\n–ë–∞—Å—Ç–∞–ø“õ—ã —Å–∞–Ω–¥–∞—Ä:", numbers)
print("–ö–≤–∞–¥—Ä–∞—Ç—Ç–∞–ª“ì–∞–Ω:", squared)

# üîπ –ú–∞—Å—Å–∏–≤—Ç–µ—Ä–º–µ–Ω –∂–∏—ñ “õ–æ–ª–¥–∞–Ω—ã–ª–∞—Ç—ã–Ω —Ñ—É–Ω–∫—Ü–∏—è–ª–∞—Ä
data = np.random.randint(1, 100, size=10)
print("\n–ö–µ–∑–¥–µ–π—Å–æ“õ –º–∞—Å—Å–∏–≤:", data)
print("–û—Ä—Ç–∞ –º”ô–Ω—ñ:", np.mean(data))
print("–ú–∞–∫—Å–∏–º—É–º:", np.max(data))
print("–ú–∏–Ω–∏–º—É–º:", np.min(data))
print("–°—Ç–∞–Ω–¥–∞—Ä—Ç—Ç—ã –∞—É—ã—Ç“õ—É:", np.std(data))
print("“ö–æ—Å—ã–Ω–¥—ã:", np.sum(data))

print("\n=== ‚úÖ –ë–∞—Ä–ª—ã“õ –±”©–ª—ñ–º–¥–µ—Ä —Å”ô—Ç—Ç—ñ –æ—Ä—ã–Ω–¥–∞–ª–¥—ã! ===")
```
