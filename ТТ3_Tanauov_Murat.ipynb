**Тапсырма 1: CSV Файлын Оқу және Тізімге Сақтау**

```python
import csv
with open('Salaries.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    data_rows = []

# Файлдың тек алғашқы 10 деректер жолын (data_rows атты тізімге) сақтаңыз. Тақырып жолын (header) санамаңыз.

    for i, row in enumerate(reader):
        if i >= 10:       # тек алғашқы 10 жол
            break
        data_rows.append(row)

# data_rows тізімінің ұзындығын (len()) және бірінші элементін (бірінші қызметкердің дерегі) экранға шығарыңыз.

print("data_rows uzyndygy:", len(data_rows))
if data_rows:
    print("Birinshi kyzmetker derekteri:", data_rows[0])
else:
    print("Fail bos, derekter jok.")
```

**Тапсырма 2: Кортеж және сөздік – алғашқы 3 адамның атын сақтау**


```python
# data_rows тізімінен алғашқы 3 қызметкердің Атын (EmployeeName) алып, оларды Кортежге (first_three_names) сақтаңыз.

first_three_names = tuple(row['EmployeeName'] for row in data_rows[:3])
employee_titles = {}

# Осы 3 қызметкердің әрқайсысы үшін аты-жөнін (EmployeeName) Кілт ретінде, ал Жұмыс Атағын (JobTitle) Мән ретінде пайдаланып, Сөздік (employee_titles) құрыңыз.

for row in data_rows[:3]:
    name = row['EmployeeName']
    title = row['JobTitle']
    employee_titles[name] = title

# Кортежді және Сөздікті экранға шығарыңыз.

print("Кортеж (алғашқы 3 ат):", first_three_names)
print("Сөздік (аты → қызмет атағы):", employee_titles)
```

**Тапсырма 3: Тізімдік Өрнектер (List Comprehension)**


```python
# data_rows тізіміндегі 10 қызметкердің барлығының Базалық Жалақысын (BasePay) мәтіндік (str) түрде алып, оларды тізімдік өрнек (List Comprehension) арқылы base_pays_str атты жаңа тізімге жинаңыз.
# Ескерту: Бұл жерде түрлендіру (float-қа) немесе сүзгілеу (filtering) қажет емес, тек мәтіндік деректі алу жеткілікті.

base_pays_str = [row['BasePay'] for row in data_rows]

# Жасалған base_pays_str тізімін экранға шығарыңыз.

print("base_pays_str:", base_pays_str)
```

**Тапсырма 4: Жиын (Set) – Бірегей Атақтарды Табу**


```python
titles_list = []
for row in data_rows:
    titles_list.append(row['JobTitle'])

# Бұл тізімді Жиын (unique_titles) құрылымына түрлендіріңіз.

unique_titles = set(titles_list)

# Жиынның барлық элементтерін экранға шығарыңыз.

print("unique_titles:", unique_titles)
```
