```python
# 1-тапсырма. numbers.txt файлын оқып, ондағы сандардың қосындысын экраға шығару
with open("numbers.txt", 'r', encoding='utf-8') as f:
    print(sum(int(line) for line in f))
```

```python
# 2-тапсырма. [1...20] диапазонында тек 3-ке бөлінетін сандардың кубтарын шығарыңыз
кубтар = [x**3 for x in range(1, 21) if x % 3 == 0]
print(кубтар)
```

```python
# 3-тапсырма. Студенттердің аты мен бағалары бар сөздік жасап, ең жоғары бағаны тап
студенттер = {
    "Арман": 85,
    "Айна": 92,
    "Марат": 78,
    "Самат": 95,
    "Тимур": 88
}
жоғары_баға = max(студенттер.values())
print("Ең жоғары баға:", жоғары_баға)
```

```python
# 4-тапсырма. Берілген тізімнен қайталанатын элементті алып тастап, жиын ретінде шығарыңыз.
numbers = [1, 2, 3, 2, 4, 1, 5, 3]
unique_numbers = set(numbers)
print(unique_numbers)
```

```python
# 5-тапсырма. JSON: Бағдарламаны жазып, студенттердің аты мен бағаларын енгізіп,
# оларды JSON файлға сақтаңыз және орташа бағаны есептеңіз.
import json

import json

# Студенттердің аты мен бағалары
students = {
    "Арман": 85,
    "Самат": 92,
    "Асан": 95
}

# JSON файлға сақтау
with open("students.json", "w", encoding="utf-8") as f:
    json.dump(students, f)

# JSON файлдан оқу
with open("students.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Орташа бағаны есептеу
average = sum(data.values()) / len(data)

print("Орташа баға:", average)
```
